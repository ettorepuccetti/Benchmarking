```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

log = pd.read_csv("log/clientInterval25932.csv", header=None, names=["Interval Time","Throughput"])
performance = log["Throughput"][:-1]
allTimes = np.array(log["Interval Time"])
sumTimes = [allTimes[0:i].sum() for i in range(1,len(allTimes)) ]

plt.figure(figsize=(15,7))
plt.grid()
plt.plot(sumTimes,performance)
plt.ylabel("Throughput")
plt.xlabel("time [s]")

plt.show()
```

```python
def sumdifferent (a, b):
    
    if len(a) < len(b):
        c = b.copy()
        c[:len(a)] += a
        c[:len(a)] /= 2
    else:
        c = a.copy()
        c[:len(b)] += b
        c[:len(b)] /= 2
    
    return c
```

```python
first = 25931
nclient = 2
sumlist = np.zeros((100,2))
for i in range(first, first + nclient):
    log = pd.read_csv("log/clientInterval"+str(i)+".csv", header=None).values[:-2]
    print (log)
    sumlist = sumdifferent(sumlist, log)

sumlist = sumlist / nclient
sumlist
```
